```python
#for loop
#A for loop is used for iterating over a sequence
# (that is either a list, a tuple, a dictionary, a set, or a string).

items = ["Mango","Apple" , "orange"]
for e in items:
    print(e)
```

```python
for count in range(10):
    print(count)
```

```python
#string
for y in "Banana":
  print(y)
```

```python
#break statement
#With the break statement we can stop the loop before it has looped through all the items:
My_list = ["Papaya","Banana","Kiran","VSCode"]
for z in My_list:
  print(z)
  if z =="Banana":
    break
```

```python
#Exit the loop when x is "banana", but this time the break comes before the print:
lists = ["kiran", "kinza","Ayesha", "VSCode"]
for b in lists:
  if b == "Ayesha":
    break
  print(b)
```

```python
#continue
st = ["kiran","kinza","ayesha","vscode"]
for d in st:
  if d == "ayesha":
    continue
  print(d)
```

```python
#range in for loop
for x in range(10):
  print(x)
```

```python
#else in for loop
for y in range(8):
  print(y)
else:
  print("finally finished")
```

```python
for x in range(2, 20,3):
  print(x)
```

```python
for y in range(10):
  if y == 5:
    break
  print(y)
else:
  print("finally finished")
```

```python
#Nested loops
lst = ["apple", "NEST", "orange"]
lst2 = ["kiran", "kinza", "ayesha"]
for x in lst:
  for y in lst2:
    print(x,y)
```

```python
#pass statement
#for loops cannot be empty, but if you for some reason have a for
# loop with no content, put in the pass statement to avoid getting an error.
for x in [2,3,5]:
  pass
```
