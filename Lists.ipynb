```python
thislist =  ["Apple" , "Banana" , "Cherry"]
print(thislist)
print(type(thislist))
print(len(thislist))    #list length
print(thislist[1])
```

```python
#allow duplicates in lists
thislist = ["Apple" , "Banana" , "Cherry" , "Apple" , "Cherry"]
print(thislist)
```

```python
#list items / Data types
list1 = ["Apple" , "Banana" , "Cherry"]
list2 = [45 ,65, 68]
list3 = [True , False , False]
print(list1)
print(list2)
print(list3)
```

```python
#list can contain different data types
thislist = ["Apple" , 65 , "Banana" , True , "Male"]
print(thislist)
```

```python
#type of my list
thislist = ["Apple" , 65 , "Banana" , True , "Male"]
print(type(thislist))
```

```python
#list constructor
thislist = list(("Apple" , 63 , "Mango"))
print(thislist)
print(type(thislist))
```
