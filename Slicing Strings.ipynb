```python
#slicing strings
b = "Hello World!"
print(b[2:5])
```

```python
#slice from the start
b = "Hello World!"
print(b[:5])
```

```python
#Slice to the end
b = "Hello World!"
print(b[2:])
```

```python
#negative indexing
b = "Hello World!"
print(b[-5:-2])
```

```python
b = "Hello World!"
print(b[-6:-1])
```
